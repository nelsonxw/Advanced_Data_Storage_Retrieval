```python
# import dependencies
import datetime as dt
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sqlalchemy
from sqlalchemy import create_engine,inspect,func
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
```

```python
# create an engine to a SQLite database file called `hawaii.sqlite`
engine = create_engine('sqlite:///hawaii.sqlite')
```

```python
# use inspector to explore the database and get a list of table names
inspector = inspect(engine)
inspector.get_table_names()
```

```python
# explore the column header in each table and get a list of column name and type
columns = inspector.get_columns('measurement')
for column in columns:
    print(column['name'],column['type'])
```

```python
# explore the column header in each table and get a list of column name and type
columns = inspector.get_columns('station')
for column in columns:
    print(column['name'],column['type'])
```

```python
# reflect Database and tables into ORM class
Base = automap_base()
Base.prepare(engine,reflect=True)
Station = Base.classes.station
Measurement = Base.classes.measurement
```

```python
# create a session
session = Session(engine)
```

```python
# since the last date point in the dataset is back in August 2017, change current date to one year prior to today
current_date = dt.date.today() - dt.timedelta(days=365)
# make the lookback period as 12 months from current date
lookback_date = current_date - dt.timedelta(days=365)
```

```python
# query the average precipitation of last 12 months, group and order by date
precipitation = session.query(Measurement.date, func.avg(Measurement.prcp))\
.filter(Measurement.date >= lookback_date)\
.filter(Measurement.date < current_date)\
.group_by(Measurement.date)\
.order_by(Measurement.date).all()
```

```python
# read query result into dataframe and make the date as index
df_precipitation = pd.DataFrame(precipitation,columns=['Date','Precipitation'])
df_precipitation.set_index('Date',inplace=True)
```

```python
# plot a line chart of the precipitation with date as the x axis
sns.set()
ax = df_precipitation.plot(figsize=(9,6),fontsize=15,grid=True,x_compat=True,linewidth=10,color='dodgerblue')
ax.legend(loc='upper left',fontsize=15,frameon=True)
ax.set_yticks([0,2,4,6])
plt.xticks(rotation='45')
plt.xlabel('Date', fontsize=18)
plt.tight_layout()
plt.show()
```

```python
# show some statiscs summary of the precipitation data
df_precipitation.describe()
```

```python
# calcuate the total number of stations in Hawaii
session.query(Measurement).group_by(Measurement.station).count()
```

```python
# list the stations and temporature observation counts in descending order
station_tobs_count = session.query(Measurement.station,func.count(Measurement.tobs))\
.group_by(Measurement.station)\
.order_by(func.count(Measurement.tobs).desc())
```

```python
# select the most activate station which has the most observation counts
top_station = station_tobs_count[0]
```

```python
# query the last 12 months' temporature of the most active station
tobs = session.query(Measurement.station,Measurement.date,Measurement.tobs)\
.filter(Measurement.date >= lookback_date)\
.filter(Measurement.date < current_date)\
.filter(Measurement.station == top_station[0])\
.order_by(Measurement.date).all()
```

```python
# read the temperatures into dataframe
df_tobs = pd.DataFrame(tobs,columns=['Station','Date','tobs'])
```

```python
# plot a histogram of the temporature with 12 bins
df_tobs.plot.hist(bins=12,figsize=(9,6),grid=True,color='dodgerblue',fontsize=15)
plt.legend(loc='upper right',fontsize=20,frameon=True)
plt.ylabel('Frequency', fontsize=20)
plt.tight_layout()
plt.show()
```

```python
# define trip start date and end date
trip_start_date = '2018-05-01'
trip_end_date = '2018-05-05'
# set the check dates as one year before trip dates
check_start_date = dt.datetime.strptime(trip_start_date,'%Y-%m-%d') - dt.timedelta(days=365)
check_end_date = dt.datetime.strptime(trip_end_date,'%Y-%m-%d') - dt.timedelta(days=365)
```

```python
def calc_temps(start_date,end_date):
    """take a start date and end date, and calculate the min max and average temporatur of that time period"""
    # query temperatures during the time period
    temp = session.query(Measurement.date,Measurement.tobs)\
    .filter(Measurement.date >= start_date)\
    .filter(Measurement.date < end_date).all()
    # read query results into dataframe
    df_temp = pd.DataFrame(temp,columns=['Date','temp'])
    # return the min max and average of the temporatures
    return df_temp.temp.min(),df_temp.temp.max(),df_temp.temp.mean()
```

```python
# assing min max and average to varialbles
min,max,mean = calc_temps(check_start_date,check_end_date)
```

```python
# plot min max and average into bar chart, with max-min as the yerr
fig, ax = plt.subplots(figsize=(3,6))
ax.bar(.5,mean,.6,color='lightsalmon',alpha = .8,yerr=max-min)
ax.grid(axis='x')
ax.set_xlim(0,1)
ax.set_xticks([])
ax.set_ylabel('Temp(F)',fontsize=15)
ax.set_ylim(0,100)
plt.yticks(fontsize=15)
plt.title('Trip Avg Temp',fontsize=15)
plt.show()
```
